## **About the dataset**

This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. The objective is to predict based on diagnostic measurements whether a patient has diabetes

1) Pregnancies : Number of times pregnant

2) Glucose : Plasma glucose concentration a 2 hour in an oral glucose tolerance test

3) Blood Pressure : Diastolic Blood pressure (mm Hg)

4) SkinThickness : Triceps skin fold thickness(mm)

5) Insulin : 2-hour serum insulin (mu U/ml)

6) DiabetesPedigreeFunction : Diabetes Pedigree Function

7) Age : Age

8) Outcome : 1 - Diabetes, 0 - No diabetes

# **Data Preprocessing**

```python
# Seeking the null values
dia_df.isnull().sum()
```

```python
dia_df.nunique()
```

```python
# detecting the number of rows that contains 0 in Insulin cloumn
(dia_df[dia_df.Insulin==0])
```

```python
# Imputing the value with mean value
dia_df.Insulin.replace(to_replace=0,value=dia_df.Insulin.mean(),inplace=True)
```

```python
# detecting the number of rows that contains 0 in Skin Thickness cloumn
dia_df[dia_df['SkinThickness']==0]
```

```python
# Imputing the value with mean value
dia_df.SkinThickness.replace(to_replace=0,value=dia_df.SkinThickness.mean(),inplace=True)
```

```python
# detecting the number of rows that contains 0 in  cloumn
dia_df[dia_df['BMI']==0]
```

```python
# Imputing the value with mean value
dia_df['BMI'].replace(to_replace=0,value=dia_df['BMI'].mean(),inplace=True)
```

# **Detecting and removing the outliers**

```python
# removing outliers in Blood pressure column using IQR
q1 = dia_df.BloodPressure.quantile(0.25)
q3 = dia_df.BloodPressure.quantile(0.75)
```

```python
IQR = q3-q1
upper = q3+1.5*IQR
lower = q1-1.5*IQR
```

```python
upper,lower
```

```python
dia_df = dia_df[(dia_df.BloodPressure>lower)&(dia_df.BloodPressure<upper)]
```

```python
dia_df
```

```python
# removing outliers in glucose cloumn using IQR
q1=dia_df.Glucose.quantile(0.25)
q3=dia_df.Glucose.quantile(0.75)
```

```python
IQR = q3-q1
upper,lower = q3+1.5*IQR, q1-1.5*IQR
upper,lower
```

```python
dia_df = dia_df[(dia_df.Glucose>lower)&(dia_df.Glucose<upper)]
```

```python
dia_df
```

# **Model Evaluation**

```python
# Seprate the feature column and target column
x = dia_df.drop('Outcome',axis=1)
y = dia_df.Outcome
```

```python
# train test split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Applying DecisionTreeClassifier Algorithm
from sklearn.tree import DecisionTreeClassifier
DTC_model=DecisionTreeClassifier()
DTC_model.fit(x_train,y_train)
```

```python
DTC_y_pred=DTC_model.predict(x_test)
DTC_y_pred
```

```python
# Checking Accuracy of the Model
accuracy_DTC = accuracy_score(y_test,DTC_y_pred)
accuracy_DTC
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
confusion_DTC = confusion_matrix(y_test,DTC_y_pred)
confusion_DTC
```

```python
# HeatMap for Confusion Matrix
sns.heatmap(confusion_matrix(y_test,DTC_y_pred),annot=True,fmt='.0f').set(title='Decision Tree Classifier Confusion Matrix',xlabel='Actual',ylabel='Predicted')
plt.show()
```

```python
# Checking Classification report of the Model to check Precision and Recall
cr_DTC = classification_report(y_test,DTC_y_pred)
print(cr_DTC)
```

```python
# Applying RandomForestClassifier Algorithm
from sklearn.ensemble import RandomForestClassifier
RFC_model = RandomForestClassifier()
RFC_model.fit(x_train,y_train)
```

```python
RFC_y_pred=RFC_model.predict(x_test)
RFC_y_pred
```

```python
# Checking Accuracy of the Model
accuracy_RFC = accuracy_score(y_test,RFC_y_pred)
accuracy_RFC
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
confusion_RFC = confusion_matrix(y_test,RFC_y_pred)
confusion_RFC
```

```python
# HeatMap for Confusion Matrix
sns.heatmap(confusion_matrix(y_test,RFC_y_pred),annot=True,fmt='.0f').set(title='Random Forest Classifier Confusion Matrix',xlabel='Actual',ylabel='Predicted')
plt.show()
```

```python
# Checking Classification report of the Model to check Precision and Recall
cr_RFC = classification_report(y_test,RFC_y_pred)
print(cr_RFC)
```

Determining the Best Model:

In this study, we evaluated three Classification models Decision Tree Classifier and Random Forest Classifier on our dataset to determine the most suitable model for predicting our target variable.

**Overall Recommendation**: Based on our evaluation, the **Random Forest Classifier** emerges as the most suitable model for predicting our target variable. It combines the strengths of decision trees with ensemble learning, resulting in superior predictive performance.
